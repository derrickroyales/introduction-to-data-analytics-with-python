```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Car=pd.read_excel('DataFiles/Car data saved.xlsx')
Car.head()
```

```python
Car.columns
```

```python
variables=['Min Price (Lakh)', 'Max Price (Lakh)', 'Range (kmpl)', 'CC',
       'Seats', 'Variants','Ex-Showroom Price', 'RTO', 'Insurance',
       'Other', 'Onroad Price']
```

```python
plt.figure(figsize=(15,12))
for i, var in enumerate (variables, start=1):
    plt.subplot(4,3,i)
    sns.scatterplot(x=Car[var], y=Car['Variants'], alpha=0.7)
    plt.title(f'Scatter plot of {var} vs Onroad Price')
    plt.xlabel(var)
    plt.ylabel('Target')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for i, var in enumerate(variables, start=1):
    plt.subplot(3, 4, i)
    sns.histplot(Car[var], bins=20, kde=True, color='grey')
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Create a boxplot of Range by Car Type
sns.boxplot(data=Car, x='Type', y='Range (kmpl)', hue='Type', palette=['Green','Blue'])
plt.title('Boxplot of Range by Car Type')
plt.show()
```

```python
columnsA=['Min Price (Lakh)', 'Max Price (Lakh)', 'Range (kmpl)', 'CC',
       'Seats', 'Variants','Ex-Showroom Price', 'RTO', 'Insurance',
       'Other']
dfSelected=Car[columnsA]
dfSelected.head()
```

```python
corr= dfSelected.corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='BuPu')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
sns.violinplot(data=Car, x='Type', y='Max Price (Lakh)', hue='Type', palette=['grey','orange'])
plt.title('Violin Plot of Maximum Price by Car Type')
plt.show()
```

```python
sns.violinplot(data=Car, x='Type', y='Range (kmpl)', hue='Type', palette=['grey','orange'])
plt.title('Violin Plot of Maximum Price by Car Type')
plt.show()
```
