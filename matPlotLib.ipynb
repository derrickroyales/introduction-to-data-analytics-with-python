```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('DataFiles/datawhoo.xlsx')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])    #convert to datetime formart

#plotting the data
plt.plot (df['Date'], df['Sales'], color='Yellow')
plt.title ('Sales over Time')
plt.xlabel ('Date')
plt.ylabel ('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(df['Product'], df['Sales'], color='turquoise')
plt.title('Sales Per Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
CarData=pd.read_excel('DataFiles/Car data saved.xlsx')
CarData.head()
```

```python
plt.figure(figsize=(28,10))
plt.scatter(CarData['Max Price (Lakh)'], CarData['Insurance'])

m,b = np.polyfit(CarData['Max Price (Lakh)'], CarData['Insurance'],1)
plt.plot(CarData['Max Price (Lakh)'], m * CarData ['Max Price (Lakh)']+b, color='black', label='Trend Line')

plt.title('Max Price (Lakh) Vs Insurance')
plt.xlabel('Max Price (Lakh)')
plt.ylabel('Insurance')
plt.grid(True)
plt.show()
```

```python
sales_data = df.groupby('Category')['Sales'].sum()
sales_data
```

```python
plt.figure(figsize=(8,8))
plt.pie(sales_data, labels=sales_data.index, autopct='%0.0f%%')
plt.title('Proportion of Sales by Category')
plt.show()
```

```python
plt.hist(df['Sales'], bins=5, color='turquoise', edgecolor='white')
plt.title('Sales Frequency')
plt.xlabel('Sales')
plt.ylabel('Frequency')
```

```python
counts, bin_edges, _ = plt.hist(df['Sales'], bins=5, color='turquoise', edgecolor='white')

# Calculate the center of each bin
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

# Plot points at the top center of each bar
plt.plot(bin_centers, counts, 'o', color='black')  # 'o' for circular markers

# # Connect the points with a line
plt.plot(bin_centers, counts, '--', color='black')  # '-' for a solid line

# Add titles and labels
plt.title('Sales Frequency')
plt.xlabel('Sales')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
plt.fill_between(df['Date'], df['Sales'], color = 'Lightgreen', alpha=0.3)
plt.plot(df['Date'], df['Sales'], color='green')
plt.title('Cumulative Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```
