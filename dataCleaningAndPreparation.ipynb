```python
q1 = fd['Sales'].quantile(0.25)
q3 = fd['Sales'].quantile(0.75)
iqr = q3-q1
print(q1,q3,iqr)
```

```python
lower_bound=q1-1.5*iqr
lower_bound
```

```python
upper_bound=q3+1.5*iqr
upper_bound
```

```python
required = fd[(fd['Sales'] >= lower_bound) & (fd['Sales'] <= upper_bound)]
print('All Sales\n', fd['Sales'].values)
print('')
print('Sales after removing outliers\n', required['Sales'].values)
```
